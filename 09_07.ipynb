```python
from transformers import pipeline

# Charger le modèle et le tokenizer
model_id = "distilbert-base-uncased-finetuned-sst-2-english"
sentiment_pipeline = pipeline("sentiment-analysis", model=model_id)
```

```python
# Exemple de phrase à analyser
phrase = "I absolutely love this product!"

# Analyser la phrase
result = sentiment_pipeline(phrase)

# Interpréter le résultat
sentiment = 1 if result[0]['label'] == 'POSITIVE' else -1

print(f"La phrase '{phrase}' a un sentiment de : {sentiment}")
```
