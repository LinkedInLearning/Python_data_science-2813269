```python
import matplotlib.pyplot as plt
import numpy as np
  
def sigmoid(x):
    return 1. / (1. + np.exp(-x))
```

```python
x = np.arange(-8, 8, 0.2)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.arange(-8, 8, 0.2)
y = tanh(x)
plt.plot(x, y)
plt.show()
```

```python
def Relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-8, 8, 0.2)
y = Relu(x)
plt.plot(x, y)
plt.show()
```

```python
def softmax(x):
    e = np.exp(x)
    return e / e.sum(axis=0)
```

```python
x = np.arange(-8, 8, 0.2)
y = softmax(x)
plt.plot(x, y)
plt.show()
```
